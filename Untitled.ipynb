```python
import os
import re
import json
import time
from pathlib import Path

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

import mlflow
import mlflow.sklearn

# Deep Learning (Keras)
import tensorflow as tf
import mlflow.tensorflow

# BERT (optionnel)
try:
    import torch
    from datasets import Dataset
    from transformers import (
        AutoTokenizer, AutoModelForSequenceClassification,
        TrainingArguments, Trainer
    )
    _BERT_AVAILABLE = True
except Exception as e:
    _BERT_AVAILABLE = False
    _BERT_IMPORT_ERROR = repr(e)

print("Imports OK | BERT disponible :", _BERT_AVAILABLE)
```

## Configuration

D√©finition des chemins, colonnes, param√®tres de split et MLflow.


```python
# BOOKMARK 1 ‚Äî Configuration

DATA_PATH = "data/tweets.csv"   # ‚úÖ ton fichier r√©el
TEXT_COL = "text"
LABEL_COL = "label"             # 1 = n√©gatif, 0 = non n√©gatif

RANDOM_STATE = 42
TEST_SIZE = 0.2
VAL_SIZE = 0.2

# === MLflow local ===
tracking_dir = os.path.abspath("mlruns")
os.makedirs(tracking_dir, exist_ok=True)
mlflow.set_tracking_uri(f"file:///{tracking_dir.replace(os.sep, '/')}")

EXPERIMENT_NAME = "airparadis_sentiment"
mlflow.set_experiment(EXPERIMENT_NAME)

print("CWD :", Path.cwd())
print("DATA_PATH :", DATA_PATH, "| Exists ?", Path(DATA_PATH).exists())
print("MLflow URI :", mlflow.get_tracking_uri())
```

##  Utils

- Nettoyage des tweets
- Calcul de m√©triques binaires
- Logging de m√©triques vers MLflow



```python
# BOOKMARK 2 ‚Äî Utils

_URL_RE = re.compile(r"http\S+|www\.\S+")
_MENTION_RE = re.compile(r"@\w+")
_HASHTAG_RE = re.compile(r"#")

def clean_tweet(text: str) -> str:
    text = str(text)
    text = _URL_RE.sub("", text)
    text = _MENTION_RE.sub("@user", text)
    text = _HASHTAG_RE.sub("", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text


def compute_metrics(y_true, y_proba):
    y_true = np.asarray(y_true).astype(int)
    y_proba = np.asarray(y_proba).astype(float)

    y_pred = (y_proba >= 0.5).astype(int)

    return {
        "accuracy": accuracy_score(y_true, y_pred),
        "f1": f1_score(y_true, y_pred),
        "precision": precision_score(y_true, y_pred, zero_division=0),
        "recall": recall_score(y_true, y_pred, zero_division=0),
        "roc_auc": roc_auc_score(y_true, y_proba),
    }

def log_metrics(metrics, prefix=""):
    for k, v in metrics.items():
        mlflow.log_metric(prefix + k, float(v))
```

## Chargement du dataset brut

Le fichier :
- est encod√© `latin-1`
- utilise `/` comme s√©parateur
- n‚Äôa pas de header fiable (`header=None`)


```python
# BOOKMARK 3 ‚Äî Load dataset (Sentiment140)

df_raw = pd.read_csv(
    DATA_PATH,
    encoding="latin-1",
    header=None
)

df_raw.columns = ["target", "id", "date", "query", "user", "text"]

print("Shape brute :", df_raw.shape)
df_raw.head()
```

## onstruction du dataset final

Sentiment140 :
- target = 0 (n√©gatif)
- target = 4 (positif)

Convention projet :
- label = 1 si n√©gatif
- label = 0 sinon


```python

df = df_raw[["text", "target"]].copy()

df[LABEL_COL] = (df["target"] == 0).astype(int)  # 1 = n√©gatif
df[TEXT_COL] = df["text"].map(clean_tweet)

df = df[[TEXT_COL, LABEL_COL]].dropna()

print("Shape finale :", df.shape)
df.head()
```

## V√©rifications rapides

- R√©partition des labels
- Longueur des tweets
- Nombre de mots


```python

display(df[LABEL_COL].value_counts(normalize=True).rename("proportion"))

df["text_len"] = df[TEXT_COL].str.len()
df["word_count"] = df[TEXT_COL].str.split().map(len)

display(df[["text_len", "word_count"]].describe())

plt.figure()
df["text_len"].hist(bins=50)
plt.title("Distribution longueur des tweets")
plt.show()
```

## Split train / val / test

D√©coupage stratifi√© :
- test = 20%
- val = 20% (sur le reste)


```python

X = df[TEXT_COL].values
y = df[LABEL_COL].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, stratify=y, random_state=RANDOM_STATE
)

val_ratio = VAL_SIZE / (1 - TEST_SIZE)
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=val_ratio, stratify=y_train, random_state=RANDOM_STATE
)

print("Train :", len(X_train), "Val :", len(X_val), "Test :", len(X_test))
```

## VERIFICATION  Equilibre des classes

On v√©rifie la proportion de tweets n√©gatifs dans :
- train
- validation
- test

Attendu : environ 0.5 si dataset √©quilibr√©.


```python

print("Train negative rate:", np.mean(y_train))
print("Val negative rate:", np.mean(y_val))
print("Test negative rate:", np.mean(y_test))
```

##  Mod√®le baseline

Approche rapide :
- TF-IDF
- R√©gression Logistique


```python

pipe = Pipeline([
    ("tfidf", TfidfVectorizer(max_features=50000, ngram_range=(1, 2))),
    ("clf", LogisticRegression(max_iter=200, class_weight="balanced"))
])

with mlflow.start_run(run_name="baseline_tfidf_logreg"):
    t0 = time.time()
    pipe.fit(X_train, y_train)
    mlflow.log_metric("train_time_sec", time.time() - t0)

    val_proba = pipe.predict_proba(X_val)[:, 1]
    test_proba = pipe.predict_proba(X_test)[:, 1]

    val_metrics = compute_metrics(y_val, val_proba)
    test_metrics = compute_metrics(y_test, test_proba)

    log_metrics(val_metrics, "val_")
    log_metrics(test_metrics, "test_")

    mlflow.sklearn.log_model(pipe, "model")

print("Baseline test metrics :", test_metrics)
```

## VERIFICATION Baseline behaviour

On v√©rifie :
- taux de pr√©diction positive
- distribution des probabilit√©s
- classification report complet

```python

from sklearn.metrics import classification_report

val_proba = pipe.predict_proba(X_val)[:, 1]
val_pred = (val_proba >= 0.5).astype(int)

print("Predicted negative rate (val):", val_pred.mean())
print("Proba min / mean / max:", val_proba.min(), val_proba.mean(), val_proba.max())

print("\nClassification report (validation):")
print(classification_report(y_val, val_pred))
```

## Mod√®le Deep Learning (Keras)

R√©seau sur mesure :
- TextVectorization int√©gr√©e
- Embedding + BiGRU
- EarlyStopping


```python
# BOOKMARK 8 ‚Äî Deep Learning Keras (MLflow) + visuel

# ---- S√©curit√© reproductibilit√© ----
tf.keras.utils.set_random_seed(RANDOM_STATE)

max_tokens = 20000
seq_len = 60
embed_dim = 100
batch_size = 64
epochs = 5
lr = 2e-4

print("‚úÖ DL config:", {
    "max_tokens": max_tokens,
    "seq_len": seq_len,
    "embed_dim": embed_dim,
    "batch_size": batch_size,
    "epochs": epochs,
    "lr": lr
})

# ---- Vectorizer ----
vectorizer = tf.keras.layers.TextVectorization(
    max_tokens=max_tokens,
    output_mode="int",
    output_sequence_length=seq_len
)

print("\n‚è≥ [1/3] Adaptation du TextVectorization sur X_train...")
t0 = time.time()
vectorizer.adapt(np.array(X_train, dtype=str))
print(f"‚úÖ Adapt termin√© en {time.time() - t0:.1f} sec")

# ---- Mod√®le ----
model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,), dtype=tf.string),
    vectorizer,
    tf.keras.layers.Embedding(max_tokens, embed_dim, mask_zero=True),
    tf.keras.layers.Bidirectional(tf.keras.layers.GRU(64)),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(1, activation="sigmoid"),
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(lr),
    loss="binary_crossentropy",
    metrics=[tf.keras.metrics.AUC(name="auc")]
)

callbacks = [
    tf.keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True),
]

# ---- MLflow run ----
with mlflow.start_run(run_name="advanced_dl_keras"):
    # Log hyperparams (utile en UI)
    mlflow.log_params({
        "max_tokens": max_tokens,
        "seq_len": seq_len,
        "embed_dim": embed_dim,
        "rnn_units": 64,
        "dropout": 0.3,
        "batch_size": batch_size,
        "epochs": epochs,
        "lr": lr,
    })

    print("\nüöÄ [2/3] Training (affichage par epoch)...")
    t0 = time.time()
    hist = model.fit(
        np.array(X_train, dtype=str), y_train,
        validation_data=(np.array(X_val, dtype=str), y_val),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=callbacks,
        verbose=2   # ‚úÖ affichage stable (1 = barre parfois invisible)
    )
    train_time = time.time() - t0
    mlflow.log_metric("train_time_sec", train_time)
    print(f"‚úÖ Training termin√© en {train_time/60:.1f} min")

    print("\nüß™ [3/3] √âvaluation val + test...")
    val_proba = model.predict(np.array(X_val, dtype=str), verbose=0).ravel()
    test_proba = model.predict(np.array(X_test, dtype=str), verbose=0).ravel()

    val_metrics = compute_metrics(y_val, val_proba)
    test_metrics = compute_metrics(y_test, test_proba)

    log_metrics(val_metrics, "val_")
    log_metrics(test_metrics, "test_")

    mlflow.tensorflow.log_model(model, "model")

print("\nDL val metrics :", val_metrics)
print("DL test metrics:", test_metrics)
```

## API ‚Äî G√©n√©ration des fichiers FastAPI + tests
On cr√©e tout le squelette de l‚ÄôAPI dans le dossier /api.


```python
from pathlib import Path

# Dossiers
Path("api/tests").mkdir(parents=True, exist_ok=True)

# main.py
main_py = r'''
from fastapi import FastAPI
from pydantic import BaseModel
import numpy as np
import tensorflow as tf
import logging

app = FastAPI(title="AirParadis Sentiment API")

logger = logging.getLogger("airparadis")
logging.basicConfig(level=logging.INFO)

MODEL_PATH = "model"
model = tf.keras.models.load_model(MODEL_PATH)

class TweetIn(BaseModel):
    text: str

class PredOut(BaseModel):
    negative_proba: float
    negative_label: int

class FeedbackIn(BaseModel):
    text: str
    negative_proba: float
    negative_label: int
    is_wrong: bool = True

@app.get("/health")
def health():
    return {"status": "ok"}

@app.post("/predict", response_model=PredOut)
def predict(payload: TweetIn):
    text = (payload.text or "").strip()
    proba = float(model.predict(np.array([text], dtype=str), verbose=0).ravel()[0])
    label = int(proba >= 0.5)
    return {"negative_proba": proba, "negative_label": label}

@app.post("/feedback")
def feedback(payload: FeedbackIn):
    logger.info(
        "USER_FEEDBACK is_wrong=%s proba=%.4f label=%s text=%s",
        payload.is_wrong, payload.negative_proba, payload.negative_label, payload.text[:300]
    )
    return {"status": "logged"}
'''
Path("api/main.py").write_text(main_py, encoding="utf-8")

# requirements.txt
req = """fastapi
uvicorn
tensorflow==2.15.1
numpy
pydantic
pytest
"""
Path("api/requirements.txt").write_text(req, encoding="utf-8")

# tests
test_py = r'''
from fastapi.testclient import TestClient
from main import app

client = TestClient(app)

def test_health():
    r = client.get("/health")
    assert r.status_code == 200
    assert r.json()["status"] == "ok"

def test_predict():
    r = client.post("/predict", json={"text": "this airline is awful"})
    assert r.status_code == 200
    data = r.json()
    assert 0.0 <= data["negative_proba"] <= 1.0
    assert data["negative_label"] in [0, 1]
'''
Path("api/tests/test_api.py").write_text(test_py, encoding="utf-8")

print("‚úÖ Fichiers cr√©√©s : api/main.py, api/requirements.txt, api/tests/test_api.py")
```
